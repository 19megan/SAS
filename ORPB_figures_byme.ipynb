```python
# This script will create similar figure of 3.13 in Esther's dissertation for Pond Branch
# Date created: 07/28/2025

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime

df = pd.read_csv('ORPB_isotope_data.csv', index_col=0, parse_dates=True)
print(df)
print(df.columns)
```

```python
plt.figure(figsize=(13, 20))
ax=[]
ax.append(plt.subplot(4, 1, 1))
ax.append(plt.subplot(4, 1, 2, sharex=ax[0]))
ax.append(plt.subplot(4, 1, 3, sharex=ax[0]))
ax.append(plt.subplot(4, 1, 4, sharex=ax[0]))

ax[0].bar(df.index, df['rainfall (mm/hr)'], label='rainfall')
ax[0].bar(df.index, df['snowfall SWE (mm/hr)'], ',', label='snowfall SWE', color='k')
ax[0].set_ylabel('precipitation (mm/hr)')
ax[0].invert_yaxis()
ax[0].twinx
ax[0].plot(df.index, df['precip 18O'], '.', label='delta 18O')
ax[0].set_ylabel('delta 18O')
ax[0].set_title('Input -> Precipitation')
ax[0].legend()

ax[1].plot(df.index, df['snowmelt (mm/hr)'], label='snowmelt')
ax[1].plot(df.index, df['discharge (mm/hr)'], label='discharge')
ax[1].set_ylabel('(mm/hr)')
ax[1].set_title('Output 1 -> Discharge')
ax[1].legend()

ax[2].scatter(df.index, df['ET (mm/hr)'], label='ET')
ax[2].set_ylabel('(mm/hr)')
ax[2].set_title('Output 2 -> ET')
ax[2].legend()

ax[3].plot(df.index, df['storage (mm)'].diff(), label='storage', lw=0.5)
# ax[3].set_yscale('log')
ax[3].set_ylabel('(mm)')
ax[3].set_title('/Delta Storage')
ax[3].legend()
```
